```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("sales_data.csv")
```

```python
profitList = df["total_profit"].tolist()
profitList
```

```python
monthList = df["month_number"].tolist()
monthList
```

```python
plt.plot(monthList, profitList)
plt.xlabel("Month Number")
plt.ylabel("Profit in Dollars")
plt.title("Profit per Month")
plt.xticks(monthList)
plt.yticks([100000,200000,300000,400000,500000])
plt.show()
```

```python
abreu_logs = pd.read_csv("abreu_logs_1.csv")
ramirez_logs = pd.read_csv("ramirez_logs_1.csv")
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
    
abreu_logs["wRC+"].rename("Abreu wRC+ Rolling Avg").rolling(6).mean().plot(ax=ax1,legend=True)
abreu_logs["wRC+_fin"].rename("Abreu wRC+ Season Final").plot(ax=ax1,legend=True)
```

```python
fig1 = plt.figure()
ax2 = fig1.add_subplot(111)

ramirez_logs["wRC+"].rename("Ramirez wRC+ Rolling Avg").rolling(6).mean().plot(ax=ax2,legend=True)
ramirez_logs["wRC+_fin"].rename("Ramirez wRC+ Season Final").plot(ax=ax2,legend=True)
```
